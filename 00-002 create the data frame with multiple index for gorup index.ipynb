```python
import pandas as pd
```

```python
names = pd.read_csv("./new data/name.csv")["name"]
names.head()
```

```python
names.info()
```

```python
for name in names:
    print(name)
```

```python
list_of_names = []
for name in names:
    for i in range(5):
        list_of_names.append(name)
list_of_names
```

```python
len(list_of_names)/5
```

```python
years = [1398, 1399, 1400, 1401, 1402]*85
years
```

```python
mul_index_data = list(zip(list_of_names, years))
index = pd.MultiIndex.from_tuples(mul_index_data, names=["Company", "Year"])
```

```python
final_df = pd.DataFrame(index=index, columns=["mean of closed price"])
```

```python
final_df.tail(30)
```

```python
final_df.to_csv("./new data/output/processed/final_df_mean_of_closed.csv")
```
