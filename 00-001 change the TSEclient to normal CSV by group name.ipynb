```python
import pandas as pd
from khayyam import *
from datetime import date, datetime
```

```python
names = pd.read_csv("./new data/name.csv")["name"]
```

```python
def change_date_to_jalali(date_str: str):
    year = int(date_str[:4])
    month = int(date_str[4:6])
    day = int(date_str[6:])
    gregorian_date = date(year, month, day)
    jalali_date = JalaliDate(gregorian_date)
    if len(str(jalali_date.month)) == 2 and len(str(jalali_date.day)) == 2:
        return f"{jalali_date.year}{jalali_date.month}{jalali_date.day}"
    elif len(str(jalali_date.month)) != 2 and len(str(jalali_date.day)) == 2:
        return f"{jalali_date.year}0{jalali_date.month}{jalali_date.day}"
    elif len(str(jalali_date.month)) == 2 and len(str(jalali_date.day)) != 2:
        return f"{jalali_date.year}{jalali_date.month}0{jalali_date.day}"
    else:
        return f"{jalali_date.year}0{jalali_date.month}0{jalali_date.day}"
```

```python
for name in names:
    print(name)
    df = pd.read_csv(f"./new data/{name}.csv", header=None, encoding='ISO-8859-1')

    # The first row contains the header; split it into columns
    header = df.iloc[0, 0].split(',')

    # The remaining rows contain the data; split each row into columns
    data = df.iloc[1:, 0].str.split(',', expand=True)

    # Assign the split header as the column names
    data.columns = header

    # Select only the desired columns
    df = data[["<DTYYYYMMDD>", "<CLOSE>"]]

    df["<DTYYYYMMDD>"] = df["<DTYYYYMMDD>"].astype('str')
    # apply the change_date_to_dash_sparated to the <DTYYYYMMDD> column
    df["<DTYYYYMMDD>"] = df["<DTYYYYMMDD>"].apply(change_date_to_jalali)
    df.rename(columns={"<CLOSE>":"ClosePrice", "<DTYYYYMMDD>":"PDate"}, inplace=True)
    df.to_csv(f"./new data/output/{name}.csv")
```
